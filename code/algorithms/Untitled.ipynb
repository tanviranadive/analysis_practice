```r
a = accidents
unique(accidents$Accident_Severity)
a$Accident_Severity = factor(accidents$Accident_Severity, levels = c(1, 2, 3),
                             labels = c("Slight", "Serious", "Fatal"))
unique(a$Accident_Severity)
```

```r
#transform Light_Conditions into daylight and darkness
unique(accidents$Light_Conditions)
a$Light_Conditions = factor(accidents$Light_Conditions, levels = c("Daylight: Street light present", "Darkness: Street lights present and lit", "Darkness: Street lighting unknown", "Darkness: Street lights present but unlit", 'Darkeness: No street lighting'),
                            #labels = c("Daylight", "Darkness", "Darkness", "Darkness")) #labels = c(1, 2, 3, 4))
                            labels = c("Daylight", "Darkness - lights lit", "Darkness - lighting unknown", "Darkness - lights unlit", "Darkness - lights unlit")) #labels = c(1, 2, 3, 4))
unique(a$Light_Conditions)
```

```r
unique(accidents$Road_Surface_Conditions)
a$Road_Surface_Conditions = factor(accidents$Road_Surface_Conditions, levels = c("Data missing or out of range", "Dry", "Wet/Damp", "Frost/Ice", "Snow", "Flood (Over 3cm of water)", "" ),
                                   labels = c("Data missing or out of range", "Dry", "Wet or damp", "Frost or ice", "Snow", "Wet or damp", "Data missing or out of range"))
unique(a$Road_Surface_Conditions)
```

```r
unique(accidents$Road_Type)
a$Road_Type = factor(accidents$Road_Type, levels = c("Unknown", "Single carriageway", "Dual carriageway", "One way street", "Roundabout", "Slip road"),
                     labels = c("Unknown", "Single carriageway", "Dual carriageway", "One way street", "Roundabout", "Slip road")) 
unique(a$Road_Type)
```

```r
#transform Special_Conditions_at_Site into none and special
unique(accidents$Special_Conditions_at_Site)
a$Special_Conditions_at_Site = factor(accidents$Special_Conditions_at_Site, levels = c('None', 'Roadworks', 'Mud', 'Auto traffic singal out', 'Ol or diesel', 'Auto traffic signal partly defective', 'Road surface defective', 'Permanent sign or marking defective or obscured'),
                                      labels = c("None", "Special", "Special", "Special", "Special", "Special", "Special", "Special")) #labels = c(0, 1, 2, 3, 4, 5, 6, 7))
unique(a$Special_Conditions_at_Site)
```

```r
unique(accidents$Time)
a$Time = sapply(strsplit(accidents$Time,":"),
                function(x) {
                  x <- as.numeric(x)
                  x[1]+x[2]/60
                }
)
# Midnight - 6am = Night; 6am - Noon = Morning; Noon - 6pm = afternoon; 6pm - Midnight = evening
a$Time = cut(a$Time, breaks=4, labels=c("Night", "Morning", "Afternoon", "Evening"))
unique(a$Time)
```

```r

library(dplyr)
library(cluster)
library(ggplot2) 


#compute the distance matrix using ‘gower’
gower_dist <- daisy(b[, -1], metric = "gower")
summary(gower_dist)
gower_mat <- as.matrix(gower_dist)
```

```r
# number of cluster PAM
sil <- c(NA)
for(i in 2:15){
  pam_fit <- pam(gower_mat, diss=TRUE, k=i)
  sil[i] <-pam_fit$silinfo$avg.width
}
```

```r
plot(1:15, sil,
     xlab = "Number of clusters",
     ylab = "Silhouette Width")
lines(1:15, sil)
```

```r
#Cluster Interpretation k=6
pam_fit <- pam(gower_dist, diss = TRUE, k = 14)

```

```r
install.packages("Rtsne")
```

```r
library(Rtsne)
# Cluster Visualization #
#########################
tsne_obj <- Rtsne(gower_dist, is_distance = TRUE)
```

```r
#visualize clusters for k=6
tsne_data <- tsne_obj$Y %>%
  data.frame() %>%
  setNames(c("X", "Y")) %>%
  mutate(cluster = factor(pam_fit$clustering),
         id = a$Accident_Index)

head(tsne_data)
colnames(tsne_data)[3] = "cluster"

ggplot(aes(x = X, y = Y), data = tsne_data) + geom_point(aes(color = cluster))
```
